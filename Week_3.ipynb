### Polar coordinates and parametric curves

$$x^2 + y^2 = r^2$$

$ x = r cos \theta $

$y = r sin \theta $

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation
```

```python
theta_space = np.linspace(0, 2*np.pi, 100)
```

```python
x = np.cos(theta_space)
y = np.sin(theta_space)
```

```python
# (0,0) (1,0) (0,1) (1,1)

plt.plot([0, 1, 0, 1], [0, 0, 1, 1], 'r.')
```

```python
plt.plot(x, y)
```

#### Rose curve

$$ r = cos(k \theta) $$

```python
k = 1
theta_space = np.linspace(0, 2*np.pi, 500)
r = np.cos(k*theta_space)
x = np.cos(k*theta_space)*np.cos(theta_space)
y = np.cos(k*theta_space)*np.sin(theta_space)

plt.plot(x,y)
```

```python
#create blank canvas
fig, ax = plt.subplots()
ax.axis([-1,1,-1,1])
l, = ax.plot([],[])

#what to plot at every frame
def animate(k):
    x = np.cos(k*theta_space)*np.cos(theta_space)
    y = np.cos(k*theta_space)*np.sin(theta_space)
    l.set_data(x, y)

ani = animation.FuncAnimation(fig, animate, frames=20)
```

```python
from IPython.display import HTML
HTML(ani.to_jshtml())
```
