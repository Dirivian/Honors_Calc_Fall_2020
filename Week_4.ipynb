## 2d functions, surface plots and paths on a sphere

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D 
```

```python
%matplotlib notebook
```

### Meshgrid

```python
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)

X, Y = np.meshgrid(x,y)
f = np.exp(-X**2-Y**2) 
plt.contourf(X, Y, f)
```

```python
plt.matshow(f)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X,Y,f, 500)
```

### Curves on a sphere

```python
Theta = np.linspace(0, 2*np.pi, 20)
Phi = np.linspace(0, np.pi, 20)

theta, phi = np.meshgrid(Theta, Phi)

x = np.cos(theta)*np.sin(phi)
y = np.sin(theta)*np.sin(phi)
z = np.cos(phi)

fig2 = plt.figure()
ax2 = plt.axes(projection='3d')
ax2.scatter3D(x, y, z)
ax2.set_axis_off()
```

```python
fig3 = plt.figure()
ax3 = plt.axes(projection='3d')
ax3.plot_wireframe(x, y, z, rcount =20, ccount =20 )

x1 = np.cos(1)*np.sin(1)
y1 = np.sin(1)*np.sin(1)
z1 = np.cos(1)
ax3.scatter3D(x1, y1, z1, c = 'red')

x2 = np.cos(5)*np.sin(5)
y2 = np.sin(5)*np.sin(5)
z2 = np.cos(5)
ax3.scatter3D(x2, y2, z2, c = 'red')
ax3.set_axis_off()
```

```python
fig4 = plt.figure()
ax4 = plt.axes(projection='3d')
ax4.plot_wireframe(x, y, z, rcount =6, ccount =6 )

x1 = np.cos(1)*np.sin(1)
y1 = np.sin(1)*np.sin(1)
z1 = np.cos(1)
ax4.scatter3D(x1, y1, z1, c = 'red')

x2 = np.cos(5)*np.sin(5)
y2 = np.sin(5)*np.sin(5)
z2 = np.cos(5)
ax4.scatter3D(x2, y2, z2, c = 'red')

t = np.linspace(0,1,100)
x_path = np.r_[np.cos(1+4*t)*np.sin(1) , np.cos(5)*np.sin(1+4*t)] 
y_path = np.r_[np.sin(1+4*t)*np.sin(1) , np.sin(5)*np.sin(1+4*t)] 
z_path = np.r_[np.cos(1+4*t-4*t) , np.cos(1+4*t)] 

ax4.plot3D(x_path, y_path, z_path, c= 'red')

ax4.set_axis_off()
```
